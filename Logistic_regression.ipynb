```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('classification_data/Social_Network_Ads.csv').values
print(dataset)
```

```python
X = dataset[:,:-1]
y = dataset[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=5)
```

```python
# Scaling
# ※ 일반적인 Logistic regression은 이론적으로 scaling이 필요 없으나
#    sklearn의 logistic regression은 linear regression의 개선된 모델이 적용되어
#    scaling된 데이터에 대해서 좋은 결과를 보여준다.
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_sc = sc.transform(X_train)
print(X_train)
print(X_train_sc)
```

### Logistic regression

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(X_train_sc,y_train)
```

```python
X_test_sc = sc.transform(X_test)
y_pred_sc = logistic.predict(X_test_sc)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,y_pred_sc)
print(acc)
```
